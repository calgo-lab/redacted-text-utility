```python

from datasets import load_dataset
import pandas as pd
import glob
```

```python

data_stem = "/Users/toroe/Workspace/redacted-text-utility/experimental_results"
dfs = {}
for x in glob.glob(data_stem + "/*"):
    experiment_name = x.split("/")[-1]
    dfs[experiment_name] = pd.read_json(x + "/results.json").mean()
mean_results = {}
for redaction_type in ['text_redacted_with_semantic_label_mask', 'text_redacted_with_random_mask', 'text_redacted_with_generic_mask', "none"]:
    mean_results[redaction_type] = {}
    for experiment_key, experiment_results in dfs.items():
        if redaction_type.replace("_", "-") in experiment_key.split("biomedbert-")[-1].split("-on-")[0]:
            on = experiment_key.split("biomedbert-")[-1].split("-on-")[-1]
            if on not in mean_results[redaction_type].keys():
                mean_results[redaction_type][on] = []
            mean_results[redaction_type][on].append(experiment_results)
aggregated_results = {}           
for k, v in mean_results.items():
    aggregated_results[k] = {}
    for sub_k, sub_v in v.items():
        aggregated_results[k][sub_k] = pd.concat(sub_v, axis=1).T.mean()
```
